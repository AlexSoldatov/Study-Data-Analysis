```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import sys
sys.path.append("..")
```

```python
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
data[:]
```

```python
data.info()
```

```python
outFile1 = open('1_week_1-2_task.txt', 'w')
```

```python
X = data.loc[:, ["Pclass", "Fare", "Age", "Sex"]]
X.head()
```

```python
X["Sex"] = X["Sex"].map({"male": 0, "female": 1})
```

```python
y = data["Survived"]
```

```python
X = X.dropna()
y = y[X.index]
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(X, y)
```

```python
feature_importances = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
feature_importances
```

```python
print(" ".join(feature_importances.head(2).index))
```

```python
print(" ".join(feature_importances.head(2).index), file=outFile1)
outFile1.close()
```
